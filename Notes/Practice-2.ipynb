```python
df = spark.sql("SELECT * FROM adventure_works_customers")
df = df.limit(10)
df.printSchema()

display(df)
```

```python
df_selected = df.select("FirstName", "LastName", "Gender", "AnnualIncome")
display(df_selected)
#or
df.select("FirstName", "LastName", "Gender", "AnnualIncome").show()
```

```python
df.select(df.columns[1:5]).show()#it will show 1 to 4th columns
#OR
df.select(df.FirstName, df.LastName, df.Gender, df.AnnualIncome).show()
          
```

```python
df_withcolumn=df.withColumn("TotalChildren", df.TotalChildren.cast("int"))
display(df_withcolumn)

df_renamed = df_withcolumn.withColumnRenamed("TotalChildren", "Total_Children")
display(df_renamed)
```

```python
from pyspark.sql.functions import regexp_replace, col

df_income_int = df.withColumn(
    "AnnualIncome",
    regexp_replace(col("AnnualIncome"), "[$,]", "").cast("int")
)
display(df_income_int)
```

```python
df=df_income_int.withColumn("AnnualIncometax",df_income_int.AnnualIncome*0.30)
display(df)
```

```python
from pyspark.sql.functions import lit

df.withColumn("Country", lit("India")).show()

```

```python
from pyspark.sql.functions import col

df=df.withColumn("TotalChildren",col("TotalChildren").cast("int"))
display(df)
```
